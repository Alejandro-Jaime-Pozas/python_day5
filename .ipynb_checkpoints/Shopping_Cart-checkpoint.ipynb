```python
'''Build a shopping cart program with the following capabilities:

1) Takes in an input
2) Stores user input into a dictionary or list
3) The User can add or delete items
4) The User can see current shopping list
5) The program Loops until user 'quits'
6) Upon quiting the program, prints out a receipt of the items with total and quantity.

ADD/REMOVE/SHOW/QUIT CAPABILITIES'''
```

```python
# Input types: 
#     ADD item name
#     ADD item quantity (int)
#     REMOVE item name
#     REMOVE item quantity (int)
#     SHOW current shopping cart items w total and quantity
#     QUIT the program
    
# =================================================================

# Output types:
#     print added item and quantity after both inputs
#     print removed item and quantity after both inputs
#     print current receipt w item names, quantities, and total price when user inputs SHOW
#     print final receipt w item names, quantities, and total price after user QUITS
# ==================================================================

# Edge Cases/Contingencies:
    
#     is there a list of available shopping items available? or can person choose random names?
#         available: need to create and show list with fixed items to choose from
#         random: need to create random price number for each item; then store price of that item
#                 for future reference
                
#     how is receipt displayed?
#         print title line
#         print each item, price, quantity in a new line
#         print end total price for all items
        
#     what if user inputs 0 for quantity?
#         tell user to print quantity > 0
        
#     what if user inputs show when list is empty?
#         print the empty list    
    
#     if user inputs invalid value for quantity:
#         while loop to ask user for number to convert to int
# ==================================================================  

# Logical Steps no Python:
    
#     1. I need to let user see which items are available to put into their cart first
#     2. need to ask user for which item they want to put in
#     3. ask user for quantity for that item
#     4. store the item and quantity to a list/dict
#     5. let user see their shopping cart while they continue shopping
#     6. let user checkout and buy products
    
# ==================================================================

# Pseudo-code:
    
#     start main program:
        
#         create global dict with food items and their prices for user to choose from
        
#         show user the dict at start
        
#         create empty dict for user shopping cart
        
#         while loop to run while user doesnt quit:
            
#             ask user if they want to ADD/REMOVE/SHOW/QUIT
    
#         use a fn for each task:

#         fn to ADD item and quantity:
#             def fn with user input ADD:
#                 ask user which item they want to add from above list
#                 if user item is in dict, add to shopping cart; else keep asking
#                 ask user qty for that item
#                 if user qty > 0, add to shopping cart; else keep asking
                
#         fn to REMOVE item and quantity
#             def fn with user input REMOVE:
#                 ask user which item they want to add from above list
#                 if user item is in dict, add to shopping cart; else keep asking
#                 ask user qty for that item
#                 if user qty > 0, add to shopping cart; else keep asking
                
#         fn to SHOW current shopping cart item, qty, price
#             def fn with user input SHOW:
#                 print(shopping cart)
        
#         fn to QUIT program
#             def fn with user input QUIT:
#                 print(shopping cart)
#                 print(Thanks for shopping with us today!)
```

```python
# fn to SHOW current shopping cart item, qty, price
#     def fn with user input SHOW:
#         print(shopping cart item name, price, qty)
#         print(total)
# sample: Cheese:  $0.99 / each  ---  x 4

def show_items():
    
    clear_output()
    total = 0
    print("\nHere is your shopping cart: \n")
    for item, qty in shopping_cart.items():
        print(f"\t{item}:  ${inventory[item]} / each  ---  x {qty}")
        total += inventory[item] * qty
    print(f"\nTotal pre-tax: ${total:.2f}")
```

```python
def quit():
    
    clear_output()
    total = 0
    print("\n\nThanks for shopping with us today!!")
    print("\n\nHere's your receipt for today: \n")
    for item, qty in shopping_cart.items():
        print(f"\t{item}:  ${inventory[item]} / each  ---  x {qty}")
        total += inventory[item] * qty
    
    print(f"\nTotal before tax: {total:.2f}")
    total_net = total * 1.08
    print(f"\n+ Sales tax: ${(total_net - total):.2f}")
    print(f"\nTotal: ${total_net:.2f}\n")
    
quit()
```

```python
# fn to ADD item and quantity:
#     def fn with user input ADD:
#         ask user which item they want to add from above list
#         if user item is in dict, add to shopping cart; else keep asking
#         ask user qty for that item
#         if user qty > 0, add to shopping cart; else keep asking

def add_items():
    
    name_item = input("\nWhich item do you want to add: ").capitalize()
    while name_item not in inventory:
        print('\nSorry we do not have that item.')
        name_item = input("\nWhich item do you want to add: ").capitalize()
    
    while True:
        try:
            qty_item = int(input(f"\nWhat quantity of {name_item} do you want: "))
            if qty_item < 1:
                print('\nSorry we need at least a qty of 1.')
                continue
            break
        except ValueError:
            print('\nSorry we need a number.')
            
    if name_item not in shopping_cart:
        shopping_cart[name_item] = qty_item
    else:
        prev_qty = shopping_cart[name_item]
        shopping_cart[name_item] = prev_qty + qty_item
        
    print(f"\nYou added {qty_item} of {name_item} to your cart!") # watch out for local vs global
        
# add_items()
```

```python
# fn to REMOVE item and quantity
#     def fn with user input REMOVE:
#         ask user which item they want to add from above list
#         if user item is in dict, add to shopping cart; else keep asking
#         ask user qty for that item
#         if user qty > 0, add to shopping cart; else keep asking

def remove_items():
    
    name_item = input("\nWhich item do you want to remove: ").capitalize()
    while name_item not in shopping_cart:
        print('\nSorry you do not have that item in your cart.')
        name_item = input("\nWhich item do you want to remove: ").capitalize()
    
    prev_qty = shopping_cart[name_item]
    
    while True:
        try:
            qty_item = int(input(f"\nWhat quantity of {name_item} do you want to remove: "))
            if qty_item < 1:
                print('\nSorry we need at least a qty of 1.')
                continue
            elif qty_item > prev_qty:
                print(f"\nSorry you only have {prev_qty} of {name_item} in your cart.")
                continue
            break
        except ValueError:
            print('\nSorry we need a number.')
            
    prev_qty = shopping_cart[name_item]
    shopping_cart[name_item] = prev_qty - qty_item
    if shopping_cart[name_item] == 0:
        del shopping_cart[name_item]
    print(f"\nYou removed {qty_item} of {name_item} from your cart!") # watch out for local vs global
        
# remove_items()
```

```python
from IPython.display import clear_output # leave clear output to the end
import pprint

# create global dict with food items and their prices for user to choose from

inventory = {
    "Soda": 0.99,
    "Milk": 1.99,
    "Chips": 3.99,
    "Eggs": 3.99,
    "Bread": 4.99,
    "Burrito": 5.99,
    "Candy": 6.99,
    "Cheese": 7.99,
    "Beer": 8.99,
    "Water": 0.99,
    "Cigarettes": 10.99,
    "Chocolate": 3.99,
    "Snack": 1.99,
    "Cookies": 3.99,
    "Apples": 0.99,
    "Oranges": 0.99,
    "Bananas": 0.99,
}


# create empty dict for user shopping cart

shopping_cart = {}

# while loop to run while user doesnt quit:

user_shopping = True

#     ask user if they want to ADD/REMOVE/SHOW/QUIT

while user_shopping:
    
#     clear_output()
    
    # show user the inventory items at start
    
    choice = input("\nWhat would you like to do? ADD/REMOVE/SHOW/QUIT: ").lower()
    
    clear_output()

    while choice not in ['add', 'remove', 'show', 'quit']:
        print("\nSorry, not a valid choice.")
        choice = input("\nWhat would you like to do? ADD/REMOVE/SHOW/QUIT: ").lower()
    
    print("\nHere is a list of available items to buy and their prices:\n")
    pprint.pprint(inventory)    
        
    if choice == 'add':
        add_items()

    if choice == 'remove':
        show_items()
        remove_items()

    if choice == 'show':
        show_items()

    if choice == 'quit':
        quit()
        break # need while loop to work

#     clear_output()
```
